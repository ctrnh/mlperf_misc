```python
google_classes_file_path = './google/class-descriptions-boxable.csv'
def load_google_class(classes_file_path):
    GOOGLE_CLASSES = {}
    GOOGLE_CLASSES_reverse = {}
    with open(classes_file_path, 'r') as csvfile:
        spamreader = csv.reader(csvfile)#, delimiter=' ', quotechar='|')
        for i,row in enumerate(spamreader):
            GOOGLE_CLASSES[row[1].lower()] = row[0]
            GOOGLE_CLASSES_reverse[row[0]] = row[1].lower()
    return GOOGLE_CLASSES, GOOGLE_CLASSES_reverse
GOOGLE_CLASSES, GOOGLE_CLASSES_reverse =load_google_class(google_classes_file_path)
list(GOOGLE_CLASSES.keys())
```

```python
import requests
url = "https://raw.githubusercontent.com/CSAILVision/sceneparsing/master/objectInfo150.csv"
list_ade20k_classes = list(map(lambda x: x.split(",")[-1], requests.get(url).text.split("\n")))[1:32]

ade20k_classes_reverse = dict(zip([i for i in range(1,32)],list_ade20k_classes))
ade20k_classes = dict(zip(list_ade20k_classes,[i for i in range(1,32)]))
```

```python
def f(DATASET_CLASSES, source_sep=", ", target_sep=";"):
    intersecting_data_class= set()
    intersecting_target = set()
    intersecting_data_idx = set()
    mapping_data_target = {}
    for data_class in DATASET_CLASSES.keys():
        for data_single_class in data_class.split(source_sep):
            for target_class in list_ade20k_classes:
                for target_single_class in target_class.split(target_sep):
                    if data_single_class.lower() == target_single_class:
                        intersecting_data_class.add(data_class)
                        intersecting_data_idx.add(DATASET_CLASSES[data_class])
                        intersecting_target.add(target_class)
                        mapping_data_target[DATASET_CLASSES[data_class]] = target_class
                    elif target_single_class in data_class:
                        print(target_single_class, "g:", data_class)
    return intersecting_data_class, intersecting_target, intersecting_data_idx, mapping_data_target
intersecting_data_class, intersecting_target, intersecting_data_idx, mapping_data_target = f(GOOGLE_CLASSES)
```

```python
# intersecting classes
def intersecting_classes(self, DATASET_CLASSES):
        # TODO: modify coco/imagenet + move to input dataset level?
        if self.type == "coco":
            intersecting_data_class= set()
            intersecting_coco = set()
            intersecting_data_idx = set()
            mapping_data_coco = {}
            for data_class in DATASET_CLASSES.keys():
                for data_single_class in data_class.split(", "):
                    for coco_class in self.COCO_CLASSES.keys():
                        for coco_single_class in coco_class.split(", "):
                            if data_single_class.lower() == coco_single_class:
                                intersecting_data_class.add(data_class)
                                intersecting_data_idx.add(DATASET_CLASSES[data_class])
                                intersecting_coco.add(coco_class)
                                mapping_data_coco[DATASET_CLASSES[data_class]] = coco_class
            return intersecting_data_class, intersecting_coco, intersecting_data_idx, mapping_data_coco
```
