```python
import os
import requests
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMAGENET_CLASSES_URL = "https://gist.githubusercontent.com/yrevar/942d3a0ac09ec9e5eb3a/raw/238f720ff059c1f82f368259d1ca4ffa5dd8f9f5/imagenet1000_clsidx_to_labels.txt"
IMAGENET_CLASSES =  {v:k for (k,v) in eval(requests.get(IMAGENET_CLASSES_URL).text).items()}
```

```python
IMAGENET_CLASSES
```

```python
path= "/home/cindy/Documents/projets/MLPerf/datasets/ADE20K_2016_07_26/images"
train_path = f"{path}/training"
val_path = f"{path}/validation"
import matplotlib.pyplot as plt
```

```python
common_classes = set()
```

```python
v=0
selected_img_path = []
in_annotations = {}
for set_path in [train_path, val_path]:
    for letter in os.listdir(set_path):
        letter_path = os.path.join(set_path, letter)
        ADE_letter_classes = os.listdir(letter_path)
        for cur_class in ADE_letter_classes:
            spaced_class = " ".join(cur_class.split("_"))
            for imagenet_class in IMAGENET_CLASSES:
                if spaced_class in imagenet_class.split(", "):
                    common_classes.add(cur_class)
                    print(spaced_class)
                    print( imagenet_class.split(", "))
                    print( IMAGENET_CLASSES[imagenet_class])
                    class_path = os.path.join(letter_path, cur_class)
                    for f in os.listdir(class_path):
                        if f.endswith(".jpg"):
                            img_path = os.path.join(class_path, f)
                            selected_img_path.append(img_path)
                            in_annotations[img_path] = IMAGENET_CLASSES[imagenet_class]
                            v+=1
```
